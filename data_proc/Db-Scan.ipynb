# Clustering 


```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('data13.csv', sep = ';')
df.head()
```

```python
df.pop("Unnamed: 0")
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
y_pred = kmeans.fit_predict(df)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.iloc[:,0],df.iloc[:,1], df.iloc[:,2],
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 200,
           c=y_pred)
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
#kmeans clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=y_pred)
plt.title("Predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(kmeans.cluster_centers_[0][0], kmeans.cluster_centers_[0][1], marker = '*', s= 100)
plt.scatter(kmeans.cluster_centers_[1][0], kmeans.cluster_centers_[1][1], marker = '*', s= 100)
plt.scatter(kmeans.cluster_centers_[2][0], kmeans.cluster_centers_[2][1], marker = '*', s= 100)
plt.show()
```

```python
elbow = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters = i)
    y_pred = kmeans.fit_predict(df)
    elbow.append(kmeans.inertia_)
```

```python
x = [i for i in range(1, 8)]
plt.plot(x, elbow);
```

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=1.8, min_samples=3)
```

```python
y_pred=clustering.fit_predict(df)
```

```python
#kmeans clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 2], c=y_pred)
plt.title("Predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
