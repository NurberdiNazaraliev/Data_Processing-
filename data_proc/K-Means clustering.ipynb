

# Cluster Analysis. K-means

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_cluster.csv', sep = ';')
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
y_pred = kmeans.fit_predict(df)
```

```python
#kmeans clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 2], c=y_pred)
plt.title("Predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### We can also take a look at the three centroids that the algorithm found:

```python
kmeans.cluster_centers_
```

```python
#kmeans clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 2], c=y_pred)
plt.title("Predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(kmeans.cluster_centers_[0][0], kmeans.cluster_centers_[0][2], marker = '*', s= 100)
plt.scatter(kmeans.cluster_centers_[1][0], kmeans.cluster_centers_[1][2], marker = '*', s= 100)
plt.scatter(kmeans.cluster_centers_[2][0], kmeans.cluster_centers_[2][2], marker = '*', s= 100)
plt.show()
```

## Elbow method

```python
elbow = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters = i)
    y_pred = kmeans.fit_predict(df)
    elbow.append(kmeans.inertia_)
```

```python
x = [i for i in range(1, 8)]
plt.plot(x, elbow);
```
